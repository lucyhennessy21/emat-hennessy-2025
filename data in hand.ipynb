# Data in Hand

## Data Collection

I collected my dataset using Spotify’s “Download My Data” feature, which provides a ZIP file containing JSON files with personal streaming information. After logging into my Spotify account, I navigated to Privacy Settings → Download your data, requested my archive, and received an email from Spotify with the link. Once extracted, the key file for this project was the StreamingHistory_music_0.json and StreamingHistory_music_1.json, which contains detailed logs of the songs I listened to, including artist names, track titles, timestamps, and duration played.

## Data Description

The dataset consists of multiple JSON files, but for this analysis I am primarily using StreamingHistory.json.
Each record represents one song played and includes the following fields:

endTime – the date and time when playback ended

artistName – the name of the artist

trackName – the title of the track

msPlayed – the number of milliseconds the track was played

Here’s what the first few rows look like after loading into a pandas DataFrame:

```python
import pandas as pd

data = pd.read_json("/Users/lucyhennessy/Desktop/Spotify Account Data/StreamingHistory_music_1.json")
data.head()
```

## Data Summary

To summarize the dataset, I examined the number of total plays, unique artists, and the overall listening time:

```python
print("Number of records:", len(data))
print("Unique artists:", data['artistName'].nunique())
print("Unique tracks:", data['trackName'].nunique())
print("Total hours listened:", data['msPlayed'].sum() / (1000*60*60))
```
